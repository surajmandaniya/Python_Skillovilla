```python
# import pandas from python library
import pandas as pd
# import numpy from python library
import numpy as np
```

```python
# Task 1: Create DataFrames and Save as CSV
# Creating Employee DataFrame
employee_df = pd.DataFrame({
    "ID": ["A001", "A002", "A003", "A004", "A005"],
    "Name": ["John Alter", "Alice Luxumberg", "Tom Sabestine", "Nina Adgra", "Amy Johny"],
    "Gender": ["M", "F", "M", "F", "F"],
    "City": ["Paris", "London", "Berlin", "Newyork", "Madrid"],
    "Age": [25, 27, 29, 31, 30]
})

# Creating Seniority DataFrame
seniority_df = pd.DataFrame({
    "ID": ["A001", "A002", "A003", "A004", "A005"],
    "Designation Level": [2, 2, 3, 2, 3]
})

# Creating Project DataFrame
project_df = pd.DataFrame({
    "ID": ["A001", "A002", "A003", "A004", "A005", "A002", "A005", "A003", "A001", "A003", "A001", "A004", "A004", "A005"],
    "Project": ["Project 1", "Project 2", "Project 3", "Project 4", "Project 5", "Project 6", "Project 7", "Project 8",
                "Project 9", "Project 10", "Project 11", "Project 12", "Project 13", "Project 14"],
    "Cost": [1002000, 2000000, 4500000, 5500000, np.nan, 680000, 400000, 350000, np.nan, 300000, 2000000, 1000000, 3000000, 200000],
    "Status": ["Finished", "Ongoing", "Finished", "Ongoing", "Finished", "Failed", "Finished", "Failed",
               "Ongoing", "Finished", "Failed", "Ongoing", "Finished", "Finished"]
})

# Saving DataFrames as CSV files
employee_df.to_csv("Employee.csv", index=False)
seniority_df.to_csv("Seniority.csv", index=False)
project_df.to_csv("Project.csv", index=False)

# printing all dataframes
employee_df, seniority_df, project_df
```

```python
# Task 2: Fill Missing Project Cost using Running Average
#Reading csv file
project_df = pd.read_csv("Project.csv")

#defining temporary variables
running_sum = 0
count = 0
#loop 
for i in range(len(project_df)):
    if not pd.isna(project_df.loc[i, "Cost"]):
        running_sum += project_df.loc[i, "Cost"]
        count += 1
    else:
        project_df.loc[i, "Cost"] = running_sum / count
#printing 
project_df
```

```python
# Task 3: Split Name into First and Last Name
#Reading csv file
employee_df = pd.read_csv("Employee.csv")
#defining variable
employee_df[['First Name', 'Last Name']] = employee_df['Name'].str.split(' ', expand=True)
employee_df.drop('Name', axis=1, inplace=True)
#Printing
employee_df
```

```python
# Task 4: Merge All DataFrames
#defining variable
final_df = employee_df.merge(seniority_df, on="ID").merge(project_df, on="ID")
#Printing
final_df
```

```python
# Task 5: Add Bonus Column
#Defining variable for adding a column
final_df['Bonus'] = np.where(final_df['Status'] == 'Finished', final_df['Cost'] * 0.05, 0)
#Printing
final_df
```

```python
# Task 6: Demotion & Removal
# Demote designation level by 1 if projects have
final_df.loc[final_df['Status'] == 'Failed', 'Designation Level'] += 1

# Remove records with designation level above 4
final_df = final_df[final_df['Designation Level'] <= 4]
#Printing
final_df
```

```python
# Task 7: Add Mr./Mrs. and Drop Gender
#Defing variables
final_df['First Name'] = np.where(final_df['Gender'] == 'M',
                                  'Mr. ' + final_df['First Name'],
                                  'Mrs. ' + final_df['First Name'])
#Removing column
final_df.drop('Gender', axis=1, inplace=True)
#Printing
final_df
```

```python
# Task 8: Promote Based on Age
#Locating Record based on condition
final_df.loc[final_df['Age'] > 29, 'Designation Level'] -= 1
#Printing
final_df
```

```python
# Task 9: Total Project Cost per Employee
#Defing variable based on condition
total_cost_df = final_df.groupby(['ID', 'First Name'])['Cost'].sum().reset_index()
#Renaming 
total_cost_df.rename(columns={'Cost': 'Total Cost'}, inplace=True)
#Printing
total_cost_df
```
